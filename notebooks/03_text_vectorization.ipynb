## One Hot Encoding

```python
import pandas as pd
sentences = ["I am Spiderman","I am not Spiderman","We are Venom"]
tokens = sorted({word.lower() for sentence in sentences for word in sentence.split()})
def one_hot(sentence):
    vector = [0] * len(tokens)
    for word in sentence.lower().split():
        if word in tokens:
            vector[tokens.index(word)] = 1
    return vector
one_hot_df = pd.DataFrame([one_hot(sentence) for sentence in sentences], columns=tokens, index=["s1","s2","s3"])
one_hot_df
```

## Bag of Words

```python
import pandas as pd

messages=pd.read_csv('../assets/SMSSpamCollection.txt', sep='\t', names=['label','message'])
```

```python
messages.head()
```

```python
import re
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus = []
for message in messages['message']:
    review = re.sub('[^a-zA-Z]', ' ', message)
    review = review.lower().split()
    review = [ps.stem(word) for word in review if word not in stop_words]
    corpus.append(' '.join(review))
len(corpus)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500)
X=cv.fit_transform(corpus).toarray()
```

# N-gram

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=100, binary=True, ngram_range=(1,2))
X=cv.fit_transform(corpus).toarray()
X
```

## TF-IDF

Term Frequency - Inverse Document Frequency

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=2500)
tfidf_matrix = tfidf.fit_transform(corpus).toarray()
feature_names = tfidf.get_feature_names_out()
pd.DataFrame(tfidf_matrix[:5, :10], columns=feature_names[:10])
```
